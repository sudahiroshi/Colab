【リストの説明】

空のリストを作る

```
list ={}
```

listに値を入れる（１）

```
list["aaa"] = "あああ"
```

listに値を入れる（２）

```
text1="bbb"
text2="いいい"
list[text1] = text2
```

指定したlistの中身を取り出す

```
list["aaa"]
```

『共通のキーを持たない時の注意』

指定したキーを待たない要素が存在するとエラーになってしまう

```
list["vvv"]
```

get(キー名)にするとキーが存在しないときNoneを返す

```
list.get("vvv")
```

使用例『リストに格納されている値と比較する』

格納されている値と比べる値（別の値でも確かめてください）

```
text_1="aaa"
text_2="あああ"
```

共通しないキーと比較可能性があるときgetをつけるとエラーにならなくなる

```
if( list.get(text_1) == text_2 ):
  print("一致しました")
else:
  print("存在しません")
```

『listの中身を表示させる』

```
for text in list:
  print(text + ":" + list[text])
```
