###forの使い方

for文で一定回数の繰り返し処理を行うにはrange()関数を用いることが多いです.

range(x):0からx-1まで繰り返す
range(x,y):xからy-1まで繰り返す

処理を5回繰り返すプログラム

```
for i in range(5):
  print(i)
```

###ifの使い方
if 条件A：

　　　処理1

else 条件B：

　　　処理2

のように使うことができます


数値を比べるプログラム

```
x=100
y=100
if x<=y:
  print("OK")
else:
    print("NO")
```
